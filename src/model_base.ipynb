```python
from torch import nn
```

```python
class ModelBase(nn.Module):
    
    def compute_all(*args, **kwargs):
        raise NotImplementedError()
```

```python
class Model(ModelBase):
    pass
```

```python
def f(a: ModelBase):
    a.compute_all(1)
```

```python
model = Model()
f(model)
```
